```python
import torch
from transformers import pipeline
import os

pipeline_dolly = pipeline(model="databricks/dolly-v2-3b",
                            torch_dtype=torch.bfloat16,
                            trust_remote_code=True,
                            device_map="auto")
```

```python
def get_completion_dolly(input):
  system = f"""
  You are an expert storyteller.
  You are good at generating long science fiction stories.
  Generate a thousand word story based on:
  """
  prompt = f"{system}\n{input}\n"
  print(prompt)
  dolly_response = pipeline_dolly(prompt,
                                  max_new_tokens=1024,
                                  min_length = 1000
                                  )
  return dolly_response[0]["generated_text"]
```

```python
prompt = "He drew his sword and pointed at the peculiar"
print(get_completion_dolly(prompt))
```

```python
prompt = "Nihara the warrior princess was calm in the face of danger. No one dared challenge her now."
print(get_completion_dolly(prompt))
```

# Falcon 7B Model

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch

# load model
model = "tiiuae/falcon-7b-instruct"
tokenizer = AutoTokenizer.from_pretrained(model)

falcon_pipeline = transformers.pipeline("text-generation",
                                        model=model,
                                        tokenizer=tokenizer,
                                        torch_dtype=torch.bfloat16,
                                        trust_remote_code=True,
                                        device_map="auto"
                                        )
```

```python

def get_completion_falcon(input):
  system = f"""
  You are an expert storyteller.
  You are good at generating long science fiction stories.
  Generate a thousand word story based on:
  """
  prompt = f"{system}\n{input}\n"
  print(prompt)
  falcon_response = falcon_pipeline(prompt,
                                    max_length=1024,
                                    do_sample=True,
                                    top_k=50,
                                    num_return_sequences=1,
                                    eos_token_id=tokenizer.eos_token_id,
                                    min_length = 1000
                                    )

  return falcon_response
```

```python
prompt = "He drew his sword and pointed at the peculiar"

response = get_completion_falcon(prompt)
print(response[0]['generated_text'])
```

```python
prompt = '''Nihara the warrior princess was calm in the face of danger.
No one dared challenge her now.
'''

response = get_completion_falcon(prompt)
print(response[0]['generated_text'])
```
